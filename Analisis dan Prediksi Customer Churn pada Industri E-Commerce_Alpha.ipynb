### Metrik Evaluasi

**Confussion Metrics:**

|                          | **Predicted Churn (1)** | **Predicted Not Churn (0)** | **Dampak Bisnis**                                   |
| ------------------------ | ----------------------- | --------------------------- | --------------------------------------------------- |
| **Actual Churn (1)**     | **True Positive (TP)**  | **False Negative (FN)**     | TP: berhasil intervensi<br>FN: kehilangan pelanggan |
| **Actual Not Churn (0)** | **False Positive (FP)** | **True Negative (TN)**      | FP: biaya retensi terbuang<br>TN: tidak ada biaya   |

* **False Positive (FP)**
Model memprediksi pelanggan akan churn, padahal sebenarnya tidak.
➝ Dampak: Perusahaan mengeluarkan biaya retensi yang sebenarnya tidak perlu.
➝ Kerugian: Customer Retention Cost (CRC) wasted

* **False Negative (FN)**
Model memprediksi pelanggan tidak churn, padahal sebenarnya churn.
➝ Dampak: Pelanggan benar-benar pergi tanpa intervensi.
➝ Kerugian: Kehilangan revenue + harus mengganti dengan pelanggan baru.
➝ Kerugian: Customer Acquisition Cost (CAC) + lost revenue

* **Customer Acquisition Cost (CAC)**
Rata-rata e-commerce India: sekitar ₹800 – ₹2,000 per customer (tergantung segmen & kompetisi)

* **Customer Retention Cost (CRC)**
Biasanya berupa diskon, voucher, promo personalisasi sekitar ₹150 – ₹500 per customer

**False Negative (FN) jauh lebih merugikan dibanding False Positive (FP)**

Karena:

FN → kehilangan pelanggan + harus bayar CAC tinggi

FP → hanya membuang biaya retensi yang relatif lebih kecil

Karena FN lebih mahal, maka model harus:

* Meminimalkan False Negative
* Mampu menangkap sebanyak mungkin pelanggan yang benar-benar churn

* **Recall (Sensitivity)**
Recall tinggi → FN rendah
→ Lebih sedikit pelanggan churn yang terlewat


Dalam proyek ini, metrik evaluasi yang dipertimbangkan:

- Recall (Primary Metric) → untuk meminimalkan False Negative (FN), mengingat biaya kehilangan pelanggan (Customer Acquisition Cost) lebih tinggi dibandingkan biaya retensi.

- Precision → untuk mengontrol False Positive (FP), sehingga campaign retensi tetap efisien dan tidak menargetkan terlalu banyak pelanggan yang sebenarnya tidak churn.

- F1-Score → sebagai keseimbangan antara Precision dan Recall untuk melihat performa model secara keseluruhan pada kelas churn.

- PR-AUC (Precision-Recall AUC) → digunakan sebagai metrik utama untuk membandingkan model pada dataset yang bersifat imbalanced (16.8% churn), karena lebih representatif dalam mengevaluasi performa model terhadap kelas positif (churn).

- ROC-AUC → untuk melihat kemampuan model secara umum dalam membedakan antara kelas churn dan non-churn (separability).

- Cross Validation → untuk memastikan stabilitas model dan mengurangi risiko overfitting melalui evaluasi pada beberapa subset data.

```python
missing_cols = df.columns[df.isnull().any()]
missing_cols
```

```python
# Uji Chi-Square: apakah pola missing berkaitan dengan variabel target (Churn)?
for col in missing_cols:
    missing_flag = df[col].isnull().astype(int)
    contingency  = pd.crosstab(missing_flag, df['Churn'])
    chi2, p, dof, expected = stats.chi2_contingency(contingency)

    result = "berkaitan dengan Churn (indikasi MAR)" if p < 0.05 else "tidak berkaitan dengan Churn (indikasi MCAR)"
    print(f"{col:<35} p={p:.4f}  →  {result}")
```

```python
# Uji T-Test: apakah pola missing pada satu kolom berkaitan dengan distribusi kolom numerik lain?
numerical_cols = df.select_dtypes(include='number').columns

for col in missing_cols:
    missing_flag = df[col].isnull().astype(int)
    significant  = []

    for num_col in numerical_cols:
        if num_col == col:
            continue
        g1 = df[missing_flag == 1][num_col]
        g2 = df[missing_flag == 0][num_col]
        if len(g1) > 0 and len(g2) > 0:
            _, p_val = stats.ttest_ind(g1, g2, nan_policy='omit')
            if p_val < 0.05:
                significant.append(f"{num_col} (p={p_val:.4f})")

    if significant:
        print(f"{col}: signifikan dengan → {', '.join(significant)}")
    else:
        print(f"{col}: tidak ada hubungan signifikan dengan kolom numerik lain")
```

```python
# Missing Indicator Model: apakah pola missing bisa diprediksi dari fitur lain?
# Jika AUC > 0.6, memperkuat klaim MAR (bukan MCAR)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler as _SS

num_features = df.select_dtypes(include='number').columns
print("Missing Indicator Model (AUC Logistic Regression):")
print(f"{'Kolom':<35} {'AUC':>6}  Interpretasi")
print("-" * 65)

for col in missing_cols:
    flag = df[col].isnull().astype(int)
    # Gunakan kolom lain yang tidak missing sebagai features
    other_cols = [c for c in num_features if c != col and df[c].isnull().sum() == 0]
    if len(other_cols) == 0:
        # Jika semua kolom punya missing, isi sementara dengan median
        X_ind = df[num_features].drop(columns=[col]).fillna(df[num_features].median())
    else:
        X_ind = df[other_cols].fillna(df[other_cols].median())

    ss = _SS()
    X_scaled = ss.fit_transform(X_ind)
    lr = LogisticRegression(max_iter=300, random_state=42)
    lr.fit(X_scaled, flag)
    auc = roc_auc_score(flag, lr.predict_proba(X_scaled)[:, 1])
    interp = "MAR (dapat diprediksi)" if auc >= 0.6 else "mendekati MCAR"
    print(f"{col:<35} {auc:>6.4f}  {interp}")
```

```python
# Masking Validation: sembunyikan 15% nilai yang diketahui, imputasi, bandingkan MAE
# Iterative Imputer vs Median Imputer
import numpy as np
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer, SimpleImputer

np.random.seed(42)
mask_ratio = 0.15

print("Masking Validation - MAE Perbandingan Iterative vs Median Imputer:")
print(f"{'Kolom':<35} {'MAE Iterative':>15} {'MAE Median':>12}  Selisih")
print("-" * 75)

num_only = df.select_dtypes(include='number').copy()

for col in missing_cols:
    # Ambil baris yang nilainya lengkap di kolom ini
    complete_idx = num_only[col].dropna().index
    n_mask = int(len(complete_idx) * mask_ratio)
    mask_idx = np.random.choice(complete_idx, size=n_mask, replace=False)

    true_values = num_only.loc[mask_idx, col].values

    # Buat salinan dengan nilai disembunyikan
    df_masked = num_only.copy()
    df_masked.loc[mask_idx, col] = np.nan

    # Iterative Imputer
    imp_iter = IterativeImputer(random_state=42, max_iter=10, initial_strategy='median')
    arr_iter = imp_iter.fit_transform(df_masked)
    df_iter   = pd.DataFrame(arr_iter, columns=num_only.columns, index=num_only.index)
    mae_iter  = np.mean(np.abs(df_iter.loc[mask_idx, col].values - true_values))

    # Median Imputer
    imp_med  = SimpleImputer(strategy='median')
    arr_med  = imp_med.fit_transform(df_masked)
    df_med   = pd.DataFrame(arr_med, columns=num_only.columns, index=num_only.index)
    mae_med  = np.mean(np.abs(df_med.loc[mask_idx, col].values - true_values))

    diff = mae_med - mae_iter
    marker = " lebih baik" if diff > 0 else " ~setara"
    print(f"{col:<35} {mae_iter:>15.4f} {mae_med:>12.4f}  {diff:+.4f}{marker}")
```

#### Analisis Mekanisme Missing Value

Tiga pendekatan digunakan untuk mengonfirmasi mekanisme missing value sebelum memilih metode imputasi:

1. **Chi-Square Test** - 6 dari 7 kolom menunjukkan hubungan signifikan antara pola missing dan variabel target Churn (p < 0.05), mengindikasikan MAR. Hanya `DaySinceLastOrder` yang tidak berkaitan (p=0.7769, mendekati MCAR).

2. **Missing Indicator Model** - Logistic Regression dilatih untuk memprediksi apakah suatu nilai akan missing berdasarkan fitur lain. Jika AUC >= 0.6, artinya pola missing dapat diprediksi dari data yang ada, memperkuat klaim MAR.

3. **Masking Validation** - 15% nilai yang diketahui disembunyikan lalu diimputasi. MAE Iterative Imputer dibandingkan dengan Median Imputer untuk memvalidasi bahwa Iterative Imputer memberikan estimasi yang lebih akurat.

Hasil ketiga analisis konsisten menunjukkan bahwa sebagian besar missing value bersifat **MAR**, sehingga Iterative Imputer yang memanfaatkan hubungan multivariat antar fitur lebih tepat dibanding imputasi sederhana. Fit dilakukan hanya pada data training untuk mencegah data leakage.

Variabel kategorikal nominal seperti `PreferredLoginDevice`, `PreferredPaymentMode`, `Gender`, `PreferedOrderCat`, dan `MaritalStatus` ditransformasikan menggunakan OneHotEncoder agar dapat diproses oleh model berbasis numerik.

**Scaling Variabel Numerikal**

Variabel numerikal distandarisasi menggunakan StandardScaler untuk menghindari dominasi fitur dengan skala besar. Scaling ini digunakan khusus untuk Logistic Regression dan KNN yang sensitif terhadap skala; model tree-based (RF, XGBoost, LightGBM) tetap menggunakan data tanpa scaling.

### Standarisasi Fitur (StandardScaler)
Fitur distandarisasi (mean=0, std=1). *Fit* hanya pada training, lalu *transform* ke test untuk mencegah *data leakage*. **Implikasi:** `X_train_scaled` / `X_test_scaled` digunakan khusus untuk Logistic Regression yang sensitif terhadap skala fitur; model tree-based tetap memakai `X_train_model`.

### 3.2.3 Train vs Test

Membandingkan F1-Score dan Recall churn antara data **training** dan **test** untuk setiap model.

Dalam konsep **Bias-Variance Tradeoff**, model yang terlalu kompleks (variance tinggi) akan "menghafal" pola training data, termasuk noise-nya, sehingga performa di data baru (test) jauh lebih rendah. Ambang 10% dipilih karena:

1. **Praktis:** Perbedaan < 10% masih dapat dikaitkan dengan variasi natural antara distribusi train dan test (sampling noise), bukan karena model benar-benar overfit.
2. **Industri:** Threshold 10% adalah konvensi umum dalam machine learning praktis, cukup sensitif untuk menangkap overfitting nyata, namun tidak terlalu ketat hingga menolak model yang sebenarnya baik.
3. **Decision Tree** sangat rentan overfitting karena tanpa pruning ia akan terus membelah node hingga setiap training sample terklasifikasi dengan sempurna (Train Recall = 1.0), sehingga gap dengan test set sangat besar.

**Interpretasi:** Gap > 0.10 = Overfit | Gap < -0.05 = Underfit | Lainnya = OK


```python
train_results = []
results = []

for name, model in model_configs.items():

    pipe = Pipeline([
        ('preprocessing', preprocessor),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)

    # ===== TEST =====
    y_pred_test = pipe.predict(X_test)
    y_prob_test = pipe.predict_proba(X_test)[:, 1]

    results.append({
        'Model': name,
        'Recall (Churn)': round(recall_score(y_test, y_pred_test), 4),
        'F1-Score (Churn)': round(f1_score(y_test, y_pred_test), 4),
        'PR-AUC': round(average_precision_score(y_test, y_prob_test), 4),
    })

    # ===== TRAIN =====
    y_pred_train = pipe.predict(X_train)
    y_prob_train = pipe.predict_proba(X_train)[:, 1]

    train_results.append({
        'Model': name,
        'Recall (Churn)': round(recall_score(y_train, y_pred_train), 4),
        'F1-Score (Churn)': round(f1_score(y_train, y_pred_train), 4),
        'PR-AUC': round(average_precision_score(y_train, y_prob_train), 4),
    })

    overfit_rows = []

for tr, te in zip(train_results, results):
    overfit_rows.append({
        'Model':             tr['Model'],
        'Train Recall':      tr['Recall (Churn)'],
        'Test Recall':       te['Recall (Churn)'],
        'Gap Recall':        round(tr['Recall (Churn)'] - te['Recall (Churn)'], 4),
        'Train F1':          tr['F1-Score (Churn)'],
        'Test F1':           te['F1-Score (Churn)'],
        'Gap F1':            round(tr['F1-Score (Churn)'] - te['F1-Score (Churn)'], 4),
        'Train PR-AUC':      tr['PR-AUC'],
        'Test PR-AUC':       te['PR-AUC'],
        'Gap PR-AUC':        round(tr['PR-AUC'] - te['PR-AUC'], 4),
    })

overfit_df = pd.DataFrame(overfit_rows)
```

```python
def status(gap):
    if gap > 0.10:
        return '⚠️ Overfit'
    elif gap < -0.05:
        return '⚠️ Underfit'
    else:
        return '✅ OK'

overfit_df['Status (F1)']     = overfit_df['Gap F1'].apply(status)
overfit_df['Status (Recall)'] = overfit_df['Gap Recall'].apply(status)

print("Gap > 0.10 = Overfit  |  Gap < -0.05 = Underfit  |  Lainnya = OK\n")

overfit_df.style.applymap(
    lambda v: 'background-color: #ffe0e0' if '⚠️' in str(v) else '',
    subset=['Status (F1)', 'Status (Recall)']
)
```

### 3.2.3 Eksperimen Imbalance Handling

Setelah baseline "polos" dievaluasi, dilakukan eksperimen penanganan class imbalance secara **terpisah dan tidak digabungkan** untuk mengetahui kontribusi masing-masing teknik.

| Teknik | Pendekatan | Mekanisme |
|--------|-----------|-----------|
| **class_weight** | Cost-sensitive | Memberikan bobot lebih tinggi pada minority class saat training |
| **SMOTE** | Oversampling | Sintesis sampel baru pada minority class via interpolasi |
| **ROS** | Oversampling | Duplikasi sampel minority class secara acak |
| **RUS** | Undersampling | Mengurangi sampel majority class secara acak |

**Model yang digunakan:** Random Forest (dipilih berdasarkan performa baseline yang paling stabil di tree-based models).

**Metrik evaluasi:** Recall sebagai primary metric, serta Precision dan F1 untuk melihat trade-off.


```python
from imblearn.pipeline import Pipeline as ImbPipeline

from sklearn.metrics import (
    recall_score, precision_score, f1_score,
    roc_auc_score, average_precision_score
)

def eval_pipeline(pipe, X_tr, X_te, y_tr, y_te, name):

    pipe.fit(X_tr, y_tr)

    y_pred = pipe.predict(X_te)
    y_prob = pipe.predict_proba(X_te)[:, 1]

    return {
        'Teknik':          name,
        'Train Size':      len(y_tr),
        'Churn Train (%)': round(y_tr.mean() * 100, 1),
        'Recall':          round(recall_score(y_te, y_pred), 4),
        'Precision':       round(precision_score(y_te, y_pred), 4),
        'F1-Score':        round(f1_score(y_te, y_pred), 4),
        'ROC-AUC':         round(roc_auc_score(y_te, y_prob), 4),
        'PR-AUC':          round(average_precision_score(y_te, y_prob), 4)
    }

resampling_results = []

# 1️⃣ Baseline
rf_baseline = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ))
])

resampling_results.append(eval_pipeline(
    rf_baseline, X_train, X_test, y_train, y_test,
    'Baseline (No Handling)'
))


# 2️⃣ class_weight='balanced'
rf_cw = Pipeline([
    ('preprocessing', preprocessor),
    ('model', RandomForestClassifier(
        n_estimators=200,
        class_weight='balanced',
        random_state=42
    ))
])

resampling_results.append(eval_pipeline(
    rf_cw, X_train, X_test, y_train, y_test,
    'class_weight=balanced'
))

rf_smote = ImbPipeline([
    ('preprocessing', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('model', RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ))
])

resampling_results.append(eval_pipeline(
    rf_smote, X_train, X_test, y_train, y_test,
    'SMOTE'
))

rf_ros = ImbPipeline([
    ('preprocessing', preprocessor),
    ('ros', RandomOverSampler(random_state=42)),
    ('model', RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ))
])

resampling_results.append(eval_pipeline(
    rf_ros, X_train, X_test, y_train, y_test,
    'ROS (Random OverSampling)'
))

rf_rus = ImbPipeline([
    ('preprocessing', preprocessor),
    ('rus', RandomUnderSampler(random_state=42)),
    ('model', RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ))
])

resampling_results.append(eval_pipeline(
    rf_rus, X_train, X_test, y_train, y_test,
    'RUS (Random UnderSampling)'
))

resampling_df = pd.DataFrame(resampling_results).sort_values('Recall', ascending=False)

print("\nPerbandingan teknik handling imbalance (Random Forest - Pipeline):")
resampling_df
```

### Interpretasi Eksperimen Resampling

**Perbandingan teknik di atas menunjukkan:**

- **Baseline (no handling):** Performa awal tanpa intervensi apapun, jadi referensi seberapa besar kontribusi teknik imbalance.
- **class_weight='balanced':** Tidak mengubah distribusi data, hanya menyesuaikan bobot saat training. Paling aman dari risiko overfitting karena tidak menciptakan data baru.
- **SMOTE:** Menghasilkan sampel sintetis pada minority class via interpolasi k-nearest neighbors. Dapat meningkatkan generalisasi dibanding ROS.
- **ROS:** Duplikasi sederhana sampel minority. Berisiko overfitting pada sampel yang diduplikasi.
- **RUS:** Mengurangi majority class. Kehilangan informasi dari data yang dibuang, namun training set menjadi lebih kecil dan seimbang.

**Rekomendasi:** Pilih teknik yang menghasilkan Recall terbaik tanpa mengorbankan Precision terlalu drastis (lihat F1-Score sebagai trade-off indicator).


### Keputusan: Tidak Menggunakan Resampling pada Model Final

Setelah mengevaluasi keempat teknik di atas, **tidak ada teknik resampling yang dibawa ke tahap tuning dan model final**. Keputusan ini didasarkan pada beberapa pertimbangan:

1. **Baseline RF sudah sangat kuat.** Random Forest baseline tanpa resampling apapun sudah menghasilkan PR-AUC > 0.98 dan Recall > 0.89 pada test set. Kontribusi tambahan dari resampling relatif marginal mengingat performa awal yang sudah tinggi.

2. **Resampling meningkatkan Recall tapi mengorbankan Precision secara signifikan.** Teknik seperti RUS dan SMOTE memang menaikkan Recall, tapi Precision turun cukup drastis, sehingga F1-Score tidak membaik secara berarti. Karena bisnis tetap membutuhkan Precision yang terkontrol (FP = biaya retensi terbuang), trade-off ini tidak menguntungkan.

3. **Threshold tuning lebih fleksibel dan reversible.** Daripada mengubah distribusi training data (yang permanen dan berisiko memperkenalkan noise sintetis), threshold tuning pada model yang sudah baik memberikan kontrol langsung antara Recall dan Precision **tanpa mengubah model itu sendiri**. Threshold bisa disesuaikan sewaktu-waktu sesuai kondisi bisnis tanpa harus retrain.

4. **Risiko overfitting lebih rendah.** Teknik seperti ROS (duplikasi) dan SMOTE (interpolasi) menambah data yang tidak ada di realita, berpotensi membuat model terlalu sensitif terhadap pola artifikal. Baseline RF + threshold tuning lebih aman dari perspektif generalisasi.

**Kesimpulan:** Strategi yang dipilih adalah **hyperparameter tuning + threshold optimization** pada ketiga kandidat model terbaik (Random Forest, XGBoost, LightGBM), bukan resampling. Threshold dipilih berdasarkan **F1 optimal** dari Out-of-Fold CV prediction, dan model final ditentukan secara dinamis berdasarkan PR-AUC tertinggi pada tahap evaluasi berikutnya.

## 3.4 Perbandingan Model Tuned + Optimasi Threshold

Ketiga model yang telah dituning (Random Forest, XGBoost, dan LightGBM) dievaluasi pada test set dengan menggunakan threshold yang dioptimalkan. Threshold default sebesar 0.5 tidak digunakan, melainkan diganti dengan threshold optimal yang diperoleh dari Out-of-Fold (OOF) cross-validation predictions pada data training untuk menghindari data leakage.

Threshold dipilih dengan memaksimalkan F1-score pada kurva Precision–Recall, sehingga diperoleh keseimbangan antara Recall dan Precision. Pendekatan ini penting dalam kasus churn prediction yang memiliki distribusi kelas tidak seimbang, di mana model perlu mendeteksi pelanggan yang berpotensi churn secara efektif tanpa menghasilkan terlalu banyak false positive.


```python
tuned_models = {
    'Random Forest': (rf_search.best_estimator_, rf_search.best_score_),
    'XGBoost':       (xgb_search.best_estimator_, xgb_search.best_score_),
    'LightGBM':      (lgb_search.best_estimator_, lgb_search.best_score_)
}

tuned_results   = []
best_thresholds = {}

for name, (model, cv_rec) in tuned_models.items():

    # ======================================================
    # Out-of-Fold prediction dari training data
    # ======================================================
    oof_prob = cross_val_predict(
        model,
        X_train,
        y_train,
        cv=skf,
        method='predict_proba',
        n_jobs=-1
    )[:, 1]

    # ======================================================
    # Cari threshold optimal
    # ======================================================
    prec_cv, rec_cv, thresh_cv = precision_recall_curve(y_train, oof_prob)

# hitung F1 score untuk setiap threshold
    f1_scores = 2 * (prec_cv[:-1] * rec_cv[:-1]) / (prec_cv[:-1] + rec_cv[:-1] + 1e-9)

# cari threshold dengan F1 terbaik
    best_idx = np.argmax(f1_scores)
    best_thresh = float(thresh_cv[best_idx])

    # ======================================================
    # Fit model ke seluruh training data
    # ======================================================
    model.fit(X_train, y_train)

    # ======================================================
    # Evaluasi pada test set
    # ======================================================
    y_prob_te  = model.predict_proba(X_test)[:, 1]
    y_pred_opt = (y_prob_te >= best_thresh).astype(int)

    tuned_results.append({
        'Model':           name,
        'CV Recall':       round(cv_rec, 4),

        'PR-AUC':          round(average_precision_score(y_test, y_prob_te), 4),
        'ROC-AUC':         round(roc_auc_score(y_test, y_prob_te), 4),

        'Recall (opt)':    round(recall_score(y_test, y_pred_opt), 4),
        'Precision (opt)': round(precision_score(y_test, y_pred_opt), 4),
        'F1 (opt)':        round(f1_score(y_test, y_pred_opt), 4),

        'Best Threshold':  round(best_thresh, 3),
    })

tuned_df = pd.DataFrame(tuned_results).sort_values('PR-AUC', ascending=False)

print("Threshold dipilih dari Out-of-Fold CV prediction\n")
tuned_df
```

### 3.5.1 Stabilitas Model - Per-Fold Evaluation (10-Fold)

Untuk membuktikan bahwa performa model final, dilakukan evaluasi 10-Fold Stratified CV pada data training menggunakan model dan threshold yang sama. Tabel menampilkan Recall, F1, dan ROC-AUC pada setiap fold beserta rata-ratanya.

Model yang baik seharusnya menunjukkan nilai yang **konsisten di semua fold**. Perbedaan antar fold yang kecil berarti model tidak bergantung pada komposisi data tertentu.


```python
kf_stable = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

fold_results = []

for fold, (tr_idx, val_idx) in enumerate(kf_stable.split(X_train, y_train), 1):

    X_fold_tr  = X_train.iloc[tr_idx]
    X_fold_val = X_train.iloc[val_idx]

    y_fold_tr  = y_train.iloc[tr_idx]
    y_fold_val = y_train.iloc[val_idx]

    # Train model
    best_model.fit(X_fold_tr, y_fold_tr)

    # ================================
    # Train Metrics
    # ================================
    y_tr_prob_f = best_model.predict_proba(X_fold_tr)[:, 1]
    y_tr_pred_f = (y_tr_prob_f >= best_thresh).astype(int)

    # ================================
    # Validation Metrics
    # ================================
    y_val_prob_f = best_model.predict_proba(X_fold_val)[:, 1]
    y_val_pred_f = (y_val_prob_f >= best_thresh).astype(int)

    fold_results.append({
        'Fold': fold,

        'Train Recall':  round(recall_score(y_fold_tr, y_tr_pred_f), 4),
        'Val Recall':    round(recall_score(y_fold_val, y_val_pred_f), 4),

        'Train F1':      round(f1_score(y_fold_tr, y_tr_pred_f), 4),
        'Val F1':        round(f1_score(y_fold_val, y_val_pred_f), 4),

        'Train ROC-AUC': round(roc_auc_score(y_fold_tr, y_tr_prob_f), 4),
        'Val ROC-AUC':   round(roc_auc_score(y_fold_val, y_val_prob_f), 4),

        'Train PR-AUC':  round(average_precision_score(y_fold_tr, y_tr_prob_f), 4),
        'Val PR-AUC':    round(average_precision_score(y_fold_val, y_val_prob_f), 4)
    })


# ======================================================
# Ringkasan Stabilitas
# ======================================================

stability_df = pd.DataFrame(fold_results)

avg_row = stability_df.drop('Fold', axis=1).mean().round(4).to_dict()
avg_row['Fold'] = 'Average'

stability_df = pd.concat(
    [stability_df, pd.DataFrame([avg_row])],
    ignore_index=True
)


print(f"Stabilitas {best_name} - 10-Fold CV (threshold={best_thresh:.3f}):")
stability_df
```

Evaluasi menggunakan 10-Fold Cross Validation menunjukkan bahwa model LightGBM memiliki performa yang stabil dengan Validation Recall rata-rata 0.885 dan ROC-AUC rata-rata 0.987. Konsistensi performa di seluruh fold menunjukkan bahwa model mampu menggeneralisasi pola churn dengan baik dan tidak bergantung pada satu pembagian data tertentu.

```python
stability_std = stability_df.drop('Fold', axis=1).std().round(4)
print("\nStd Deviation per metric:")
print(stability_std)
```

## 3.6 Confusion Matrix & SHAP Values

SHAP (SHapley Additive exPlanations) menggunakan konsep Shapley values dari game theory untuk mendistribusikan kontribusi setiap fitur secara adil terhadap setiap prediksi individual.


```python
# ===============================
# FINAL FIT (WAJIB sebelum evaluasi & SHAP)
# ===============================
best_model.fit(X_train, y_train)

y_prob_test = best_model.predict_proba(X_test)[:, 1]
y_pred_test = (y_prob_test >= best_thresh).astype(int)

# ===============================
# Confusion Matrix
# ===============================
cm = confusion_matrix(y_test, y_pred_test)

fig, ax = plt.subplots(figsize=(6, 5))
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=['Not Churn', 'Churn']
)
disp.plot(ax=ax, colorbar=False)

ax.set_title(
    f'Confusion Matrix - {best_name} (threshold={best_thresh:.3f})',
    fontsize=13, pad=12
)

plt.tight_layout()
plt.show()


# ===============================
# SHAP EXPLANATION
# ===============================

# Ambil preprocessing dan model dari pipeline
preprocessor = best_model.named_steps['preprocessing']
model_core   = best_model.named_steps['model']

# Transform data untuk SHAP
X_train_trans = preprocessor.transform(X_train)
X_test_trans  = preprocessor.transform(X_test)

# Ambil nama fitur setelah preprocessing
feature_names = preprocessor.get_feature_names_out()

X_test_trans_df = pd.DataFrame(
    X_test_trans,
    columns=feature_names
)

# ===============================
# Tree Models (RF, XGB, LGBM)
# ===============================
if hasattr(model_core, 'feature_importances_'):

    explainer = shap.TreeExplainer(
        model_core,
        feature_perturbation="interventional"
    )

    shap_values = explainer.shap_values(X_test_trans)

    if isinstance(shap_values, list):
        sv = shap_values[1]
    elif hasattr(shap_values, 'ndim') and shap_values.ndim == 3:
        sv = shap_values[:, :, 1]
    else:
        sv = shap_values

    # ===== Beeswarm Plot =====
    shap.summary_plot(
        sv,
        X_test_trans_df,
        plot_type='dot',
        max_display=15,
        show=False
    )

    plt.gcf().set_size_inches(12, 7)
    plt.title(
        f'SHAP Feature Impact (Churn) - {best_name}',
        fontsize=13,
        pad=10
    )
    plt.xlabel('SHAP value (impact on churn probability)', fontsize=11)
    plt.subplots_adjust(left=0.22, right=0.92, top=0.92, bottom=0.10)
    plt.show()

    # ===== Bar Plot =====
    shap.summary_plot(
        sv,
        X_test_trans_df,
        plot_type='bar',
        max_display=15,
        show=False
    )

    plt.gcf().set_size_inches(10, 6)
    plt.title(
        f'SHAP Feature Importance - Mean |SHAP| ({best_name})',
        fontsize=13,
        pad=10
    )
    plt.xlabel('Mean |SHAP value|', fontsize=11)
    plt.subplots_adjust(left=0.22, right=0.90, top=0.92, bottom=0.10)
    plt.show()

else:

    # ===============================
    # KernelExplainer (Non-tree model)
    # ===============================
    print(f"Best model adalah {best_name} - menggunakan SHAP KernelExplainer")

    background = shap.kmeans(X_train_trans, 50)

    explainer = shap.KernelExplainer(
        model_core.predict_proba,
        background
    )

    shap_values = explainer.shap_values(
        X_test_trans[:200],
        nsamples=100
    )

    sv = (
        shap_values[1]
        if isinstance(shap_values, list)
        else shap_values[:, :, 1]
    )

    shap.summary_plot(
        sv,
        X_test_trans_df.iloc[:200],
        plot_type='dot',
        max_display=15,
        show=False
    )

    plt.gcf().set_size_inches(12, 7)
    plt.title(f'SHAP Feature Impact - {best_name}')
    plt.subplots_adjust(left=0.22, right=0.92, top=0.92, bottom=0.10)
    plt.show()
```

Kasarnya: lebih baik kasih voucher ke orang yang tidak butuh, daripada kehilangan pelanggan tanpa sempat berbuat apa-apa.

### Confusion Matrix Pada Konteks ini:

- **True Positive (TP):** Pelanggan yang memang akan churn dan berhasil kita deteksi - ini yang kita kejar, karena mereka bisa kita selamatkan sebelum benar-benar pergi
- **False Negative (FN):** Pelanggan yang akan churn tapi luput dari radar model - ini yang paling kita hindari, karena artinya kita harus cari pelanggan baru yang biayanya 10x lebih mahal
- **False Positive (FP):** Pelanggan yang diprediksi akan churn padahal tidak - mereka tetap dapat voucher/diskon, sedikit boros tapi tidak fatal
- **True Negative (TN):** Pelanggan yang aman dan diprediksi aman - tidak perlu tindakan apapun

Intinya, selama FN rendah, kita sudah di jalur yang benar. Cost simulation di bawah ini akan membuktikan seberapa besar penghematannya.


```python
cm = confusion_matrix(y_test, y_pred_test)

TN, FP, FN, TP = cm.ravel()

fig, ax = plt.subplots(figsize=(6,5))
ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=['Not Churn','Churn']
).plot(ax=ax, colorbar=False)

ax.set_title(
    f'Confusion Matrix - {best_name} (threshold={best_thresh:.3f})',
    fontsize=13, pad=12
)

plt.tight_layout()
plt.show()


# ==============================
# PARAMETER BIAYA
# ==============================

biaya_retensi  = 300     # Cost Retention Campaign (CRC)
biaya_akuisisi = 1500    # Cost Acquisition Customer (CAC)

total_customers = cm.sum()
churn_actual = TP + FN

print("\nANALISIS 3 SKENARIO COST-BENEFIT\n")
print(f"Total Pelanggan : {total_customers}")
print(f"Actual Churn    : {churn_actual} ({churn_actual/total_customers*100:.1f}%)")
print("-"*60)


# ==============================
# SKENARIO 1 : NO ACTION
# ==============================

cost_no_action = churn_actual * biaya_akuisisi

print("1️⃣ NO ACTION (Tanpa Intervensi)")
print(f"   Semua churn hilang")
print(f"   Biaya akuisisi : {churn_actual} x {biaya_akuisisi} = {cost_no_action:,.0f}")
print()


# ==============================
# SKENARIO 2 : RETENSI MASSAL
# ==============================

cost_mass_retention = (total_customers * biaya_retensi) + (churn_actual * biaya_akuisisi)

print("2️⃣ RETENSI MASSAL (Semua pelanggan diberi promo)")
print(f"   Biaya retensi : {total_customers} x {biaya_retensi} = {total_customers*biaya_retensi:,.0f}")
print(f"   Biaya akuisisi churn tetap terjadi = {churn_actual*biaya_akuisisi:,.0f}")
print(f"   Total biaya = {cost_mass_retention:,.0f}")
print()


# ==============================
# SKENARIO 3 : MODEL TARGETED
# ==============================

cost_model = ((TP + FP) * biaya_retensi) + (FN * biaya_akuisisi)

print("3️⃣ RETENSI TARGETED (Dengan Model)")
print(f"   Biaya retensi (TP+FP) : ({TP}+{FP}) x {biaya_retensi} = {(TP+FP)*biaya_retensi:,.0f}")
print(f"   Biaya akuisisi dari FN : {FN} x {biaya_akuisisi} = {FN*biaya_akuisisi:,.0f}")
print(f"   Total biaya = {cost_model:,.0f}")
print()

print("="*60)


# ==============================
# PERBANDINGAN PENGHEMATAN
# ==============================

save_vs_no_action = cost_no_action - cost_model
save_vs_mass = cost_mass_retention - cost_model

pct_vs_no_action = (save_vs_no_action / cost_no_action) * 100
pct_vs_mass = (save_vs_mass / cost_mass_retention) * 100

print("PERBANDINGAN PENGHEMATAN")
print(f"Dengan Model vs No Action      : {save_vs_no_action:,.0f} ({pct_vs_no_action:.1f}%)")
print(f"Dengan Model vs Retensi Massal : {save_vs_mass:,.0f} ({pct_vs_mass:.1f}%)")
```
